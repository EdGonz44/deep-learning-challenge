## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns = ['EIN', 'NAME'])
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = ['T9', 'T13', 'T12', 'T2', 'T25', 'T14', 'T29', 'T15', 'T17']
# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
application_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
application_df['CLASSIFICATION'].value_counts().loc[lambda x : x>1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = ['C7000', 'C1700', 'C4000', 'C5000', 'C1270', 'C2700',
                              'C2800', 'C7100', 'C1300', 'C1280', 'C1230', 'C1400',
                              'C7200', 'C2300', 'C1240', 'C1250', 'C8000', 'C6000',
                              'C1245', 'C1800', 'C1235', 'C1237', 'C7120', 'C1500',
                              'C1820', 'C1900', 'C1256', 'C3700', 'C2561', 'C1580',
                              'C1248', 'C2500', 'C1570', 'C1257', 'C2190', 'C1720',
                              'C4100', 'C2600', 'C2710', 'C1260', 'C2380', 'C1246',
                              'C1267', 'C2150', 'C4120', 'C8210', 'C2561', 'C4500', 'C2150',
                              'C2570', 'C1580', 'C3700', 'C8200', 'C1278', 'C1238', 'C7210',
                              'C2400', 'C1600', 'C0', 'C3200', 'C1234', 'C4200', 'C1370',
                              'C2170', 'C1728', 'C1732', 'C1283', 'C1236', 'C5200', 'C6100']
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df.nunique()
```

```python
# Check which columns have very skewed data, so as to determine if possible to drop (optimization strategy)
print(application_df['SPECIAL_CONSIDERATIONS'].value_counts())
print(application_df['STATUS'].value_counts())
print(application_df['AFFILIATION'].value_counts())
```

```python
print(application_df['USE_CASE'].value_counts())
print(application_df['ORGANIZATION'].value_counts())
```

The columns for SPECIAL_CONSIDERATIONS and STATUS are heavily skewed to one side, of possible values. It may be prudent to drop these for optimization, as losing any row that has the rare value for these two columns won't actually make a large impact on the size of the dataset.
First optimiziation attempt: removing these columns.

```python
# Removing SPECIAL_CONSIDERATIONS and STATUS columns due to possible lack of impact on dataset.
application_df = application_df.drop(columns = ['SPECIAL_CONSIDERATIONS', 'STATUS'])
```

Consider changing the grouping in AFFILIATION and USE_CASE, by slightly changing their requirements. In AFFILIATION the groups family/parent, national, regional, and other are fairly small portions of the dataset, perhaps placing them as an 'Other' category may be useful. In a similar vein, perhaps in USE_CASE it would be helpful to place CommunityServ, Heathcare, and other in the same group since they are similar and make up smaller portions of the dataset. Second optimization attempt: Condensing these features.
Optimization Improvement: 0.7295

```python
# Choose a cutoff value and create a list of affiliations to be replaced
# use the variable name `affiliations_to_replace`
affiliations_to_replace = ['Family/Parent', 'National', 'Regional', 'Other']

# Replace in dataframe
for afl in affiliations_to_replace:
    application_df['AFFILIATION'] = application_df['AFFILIATION'].replace(afl,"Other")

# Check to make sure replacement was successful
print(application_df['AFFILIATION'].value_counts())
```

```python
# Choose a cutoff value and create a list of uses to be replaced
# use the variable name `use_to_replace`
use_to_replace = ['CommunityServ', 'Heathcare', 'Other']

# Replace in dataframe
for utr in use_to_replace:
    application_df['USE_CASE'] = application_df['USE_CASE'].replace(utr,"Other")

# Check to make sure replacement was successful
print(application_df['USE_CASE'].value_counts())
```

Added another feature condensing for ORGANIZATION, may not have a significant impact, but felt necessary to do so as it fell into the same concept of the changes made to the other features (multiple columns that were significantly smaller than the other columns).

```python
# Choose a cutoff value and create a list of organizations to be replaced
# use the variable name `use_to_replace`
organization_to_replace = ['Co-operative','Corporation']

# Replace in dataframe
for otr in organization_to_replace:
    application_df['ORGANIZATION'] = application_df['ORGANIZATION'].replace(otr,"Other")

# Check to make sure replacement was successful
print(application_df['ORGANIZATION'].value_counts())
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df, dtype=float)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df['IS_SUCCESSFUL'].values
X = application_df.drop('IS_SUCCESSFUL', axis=1).values
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

Third Attempt at optimization: Adding 20 neurons to both hidden layers, 100 and 50 new total in each hidden layer. Result: no significant change.

Fourth Attempt: Adding a hidden layer. Result: 0.7295, no significant change.

Fifth attempt: changing the activation functions for second and fourth layers to leaky relu from relu Result: accuracy: 0.7287 - loss: 0.6059, possible increase in loss from baseline

6th Attempt: alternating activation function from tanh to relu
Result: accuracy: 0.7299 - loss: 0.5876 fixed the loss problem slightly, but did not greatly increase accuracy

7th attempt: alternate between elu and relu
Result: accuracy: 0.7291 - loss: 0.5792 further fixed the loss effort, but maintained the accuracy at a range similar to the baseline.

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  100
hidden_nodes_layer2 = 50
hidden_nodes_layer3 = 50  # New layer with 50 neurons
hidden_nodes_layer4 = 50  # Another new layer with 50 neurons

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="elu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Third hidden layer (new)
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="elu"))

# Fourth hidden layer (new)
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer4, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

8th Attempt: will try changing the epochs by 100 while using the alternating elu relu activation factors.
Result: accuracy: 0.7290 - loss: 0.6047 Loss was increased, and accuracy was maintained.

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=200)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
